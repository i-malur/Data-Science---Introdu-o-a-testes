# TMDB

Colunas

* Budget - gasto na produção do filme  
* genres (gêneros),  
* homepage, a página do filme na internet,  
* id, um número identificador,  
* keywords, as palavras-chave associadas ao filme,  
* original_language (língua original),  
* original_title (título original),  
* overview, uma descrição básica do filme,  
* popularity, uma espécie de "nota" de popularidade calculada pelo próprio TMDB,  
* production_companies, as empresas envolvidas na produção,  
* production_contries, os países em que o filme foi produzido,  
* release_date (data de lançamento),  
* renevue (faturamento),  
* runtime, o tempo de duração (em minutos),  
* spoken_languages, as línguas faladas no filme,  
* status, se o filme foi lançado ou não,  
* tagline, uma rápida chamada do filme (como encontramos em propagandas),  
* title (título),  
* vote_average, uma nota média do filme,
* vote_count, que mostra o número de notas atribuídas ao filme.

Z-test:

Usado quando:
Amostra é grande (geralmente acima de 30 observações)  
Variância da população é conhecida  
Compara a média da amostra com uma média populacional conhecida ou hipotética  
Assume que a distribuição da amostra segue uma distribuição normal    

T-test:

Usado quando:  
Amostra é pequena (geralmente abaixo de 30 observações)  
Variância da população é desconhecida  
Compara a média de uma amostra com uma média populacional conhecida ou hipotética  
Assume que a distribuição da amostra segue uma distribuição t de Student  


Diferenças entre Z-test e T-test:

Z-test assume variância da população conhecida  
T-test assume variância da população desconhecida e deve ser estimada a partir da amostra  


Ambos os testes:

Ajudam a determinar se as diferenças observadas entre as médias são estatisticamente significativas ou se podem ter ocorrido por mero acaso.

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
  media = temp[0:i].mean()
  stats, p = ztest(temp[0:i], value = 3.4320503405352603)
  return (i, media, p)


valores = np.array([calcula_teste(i) for i in range(2, len(temp))])
plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
plt.hlines(y = 0.05, xmin = 2, xmax = len(temp), colors = 'r')
```
